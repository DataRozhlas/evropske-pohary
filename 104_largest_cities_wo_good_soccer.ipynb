```python
import os
import pandas as pd
```

```python
cities = pd.read_csv(os.path.join("data","List_of_European_cities_by_population_within_city_limits_1.csv")).drop(columns=['Date'])
df = pd.read_csv(os.path.join("data","matches.csv"))
```

```python
cities['Official\npopulation'] = pd.to_numeric(cities['Official\npopulation'].apply(lambda x: x.replace(',','')))
```

```python
cities
```

```python
df.loc[df['location'] == 'Bazaly', 'location'] = 'Bazaly, Ostrava'
```

```python
df['competition'].drop_duplicates().to_list()
```

```python
df = df[df['competition'].isin(['UCL','CWC','EL','UECL'])]
```

```python
df['city'] = df['location'].apply(lambda x: str(x).split(',')[-1].strip())
```

```python
def join_dates(row):
    if row['date'] != None:
        return row['date']
    else:
        return row['estimatedDate']
```

```python
df['date'] = df.apply(lambda row: join_dates(row), axis = 1)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
soccer_cities = df.groupby('city')['date'].max()
soccer_cities = pd.DataFrame(soccer_cities).reset_index().rename(columns={'city': 'City'})
```

```python
cities
```

```python
merg = cities.merge(soccer_cities, on='City', how='outer')
```

```python
merg = merg[merg['Official\npopulation'].notnull()]
```

```python
merg[merg['date'].isnull()].sort_values(by='Official\npopulation',ascending=False)
```
