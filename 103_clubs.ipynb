```python
import os
import pandas as pd
```

```python
df = pd.read_json(os.path.join('data','teams.json'))
```

```python
df.groupby('teamName').first().sort_values(by='founded').dropna(subset='founded').tail(10)
```

```python
def weed(x):
    x = str(x).replace('(','').replace(')','').replace('.','')
    if len(x) > 2:
        if x[-1] == 's':
            x = x[:-1]
    return x
```

```python
df['allWords'] = df['teamName'].apply(lambda x: x.split(' '))
```

```python
allNicknames = df.explode('nicknames')
```

```python
allNicknames
```

```python
allNicknames['allNicknames'] = allNicknames['nicknames'].apply(lambda x: weed(x))
```

```python
allNicknames['allNicknameWords'] = allNicknames['allNicknames'].apply(lambda x: x.split(' '))
```

```python
allNicknames = allNicknames.explode('allNicknameWords')
```

```python
allNicknames[allNicknames['nicknames'].str.lower().str.contains('miner',na=False)].groupby('teamName').first()
```

```python
allNicknames[allNicknames['nicknames'].str.lower().str.contains('white',na=False)].groupby('teamName').first()
```

```python
nicknames = allNicknames.groupby('allNicknameWords').size().nlargest(1000)
nicknames.head(60)
```

```python
allwords = df.explode('allWords')
```

```python
allwords['allWords'] = allwords['allWords'].apply(lambda x: weed(x))
```

```python
allwords.groupby('allWords').size().nlargest(50)
```
