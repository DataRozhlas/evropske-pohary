```python
import pandas as pd
```

```python
df = pd.read_csv('data/matches.csv')
```

```python
df['homeTeamUrl'].nunique()
```

```python
def goal_difference(x):
    try:
        return abs(int(x.split("(")[0].strip().split('-')[0]) - int(x.split("(")[0].strip().split('-')[1]))
    except:
        return None
```

```python
df['attendance'] = pd.to_numeric(df['attendance'], errors='coerce')
```

```python
df['goalDifference'] = df['score'].apply(lambda x: goal_difference(x))
```

```python
df[df['goalDifference'] >= 10].sort_values(by='date')
```

```python
df.sort_values(by='goalDifference',ascending=False).head(40)
```

```python
df.sort_values(by='attendance',ascending=False).head(40)
```
